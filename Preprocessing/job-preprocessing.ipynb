```python
import re
```

```python
# Filter Job title - which is post from bdjobs.com
pattern=r'-\s*For|\(For'
def filter_job_title(dataframe):
    job_title=[]
    company_name=[]
    for index,row in dataframe.iterrows():
        data=row['Job_title']
        entry=re.split(pattern, data)
        if entry[1:]:
            job_title.append(entry[0])
            company_name.append(entry[1] )
        else:
            job_title.append(entry[0])
            company_name.append(row['Company_name'])
    dataframe['Job_title']=job_title
    dataframe['Company_name']=company_name
```

```python
filter_job_title(df)
df.head()
```

```python
# Filter Company Name -remove job ID
pattern=r'\)|\(|-\s*Job'
def filter_company(dataframe):
    name=[]
    for data in dataframe['Company_name']:
        name.append( re.split(pattern,data)[0] )
    dataframe['Company_name']=name
```

```python
filter_company(df)
df.head(5)
```

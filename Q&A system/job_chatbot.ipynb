```python
# Environment Secret key
from google.colab import userdata
secr_key=userdata.get('HF_TOKEN')
```

```python
import os
from langchain_huggingface import HuggingFaceEndpoint
import warnings
warnings.filterwarnings("ignore")


# Set the Hugging Face API token
os.environ["HUGGINGFACEHUB_API_TOKEN"] = secr_key  # Replace with your API key

# Extract relevant context from the dataset
context = "\n".join(
    df.sample(10).apply(lambda x: f"Title: {x['Job_title']}, Company: {x['Company_name']}, Location: {x['Location']}, Description: {x['Description']}", axis=1)
)
# Define the Hugging Face model
repo_id = "mistralai/Mistral-7B-v0.1"  # Example model
task = "text-generation"  # Use text generation for better prompting

# Create the LLM object
llm = HuggingFaceEndpoint(repo_id=repo_id, task=task, max_length=256, temperature=0.7, token=secr_key)

# Define a function to generate answers
def answer_question(question: str):
    # Construct the prompt
    prompt = f"""You are a helpful AI assistant answering questions based on a given context.
    Use the information in the context to provide an accurate answer.

    Context: {context}

    Question: {question}

    Answer:"""

    # Invoke the model with the structured prompt
    answer = llm.invoke(prompt)
    return answer.strip()
```

```python
# Example question
question = "is there available any job for junior software developer?"
answer = answer_question(question)

print(f"Question: {question}")
print(f"Answer: {answer}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.schema import Document
import pandas as pd
from langchain.chains import RetrievalQA, LLMChain
from langchain.chains import StuffDocumentsChain  # Corrected import
from langchain.prompts import PromptTemplate
from langchain_community.llms import HuggingFaceHub  # Correct import

# Load dataset
df = pd.read_csv("jobs.csv")  # Ensure the dataset is properly formatted

# Convert dataframe into LangChain Documents
documents = [
    Document(page_content=f"Title: {row['Job_title']}, Company: {row['Company_name']}, "
                          f"Location: {row['Location']}, Description: {row['Description']}")
    for _, row in df.iterrows()
]

# Use a text splitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
split_docs = text_splitter.split_documents(documents)

# Load Hugging Face embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Store in FAISS
vectorstore = FAISS.from_documents(split_docs, embeddings)

# Create a retriever with similarity search
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 1})

# Load the Hugging Face model
llm = HuggingFaceHub(
    repo_id="mistralai/Mistral-7B-v0.1",
    model_kwargs={"temperature": 0.7, "max_length": 256},
)

# Define a prompt template for formatting the documents
document_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""You are an AI assistant that answers questions based on the given context.
    If there is a relevant job opening, provide an answer in the format:

    'Yes, there is a job opening for [Job Title] at [Company Name] in [Location]. [Job Details].'

    If no relevant job is found, say:
    'No relevant job opening found in the provided context.'

    Context: {context}

    Question: {question}

    Answer:
    """
)

# Create a document chain
'''document_chain = StuffDocumentsChain(
    llm_chain=LLMChain(llm=llm, prompt=document_prompt),
    document_variable_name="context"
)

# Create the QA system
qa_chain = RetrievalQA(combine_documents_chain=document_chain, retriever=retriever)'''

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    chain_type="stuff",  # Alternative: Use "map_reduce" or "refine"
    chain_type_kwargs={"prompt": document_prompt}
)

def clean_response(response):
    """Extracts the answer from the model output."""
    if "Answer:" in response:
        response = response.split("Answer:")[-1].strip()
    return response.strip("'").strip('"')

# Modify the answer function
def answer_question(question: str):
    result = qa_chain.run(question)
    return clean_response(result)
```

```python
# Example usage
question = "Are there any remote job openings for Python developers?"
answer = answer_question(question)
print("Q:", question)
print("A:", answer)
```

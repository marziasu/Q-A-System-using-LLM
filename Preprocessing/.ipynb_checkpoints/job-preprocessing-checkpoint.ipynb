```python
# Filter Job title - which is post from bdjobs.com
pattern=r'-\s*For|\(For'
def filter_job_title(dataframe):
    job_title=[]
    company_name=[]
    for index,row in dataframe.iterrows():
        data=row['Job_title']
        entry=re.split(pattern, data)
        if entry[1:]:
            job_title.append(entry[0])
            company_name.append(entry[1] )
        else:
            job_title.append(entry[0])
            company_name.append(row['Company_name'])
    dataframe['Job_title']=job_title
    dataframe['Company_name']=company_name
```

```python
filter_job_title(df)
df.head()
```

```python
# Filter Company Name -remove job ID
pattern=r'\)|\(|-\s*Job'
def filter_company(dataframe):
    name=[]
    for data in dataframe['Company_name']:
        name.append( re.split(pattern,data)[0] )
    dataframe['Company_name']=name
```

# job title preprocessing

```python
df['Job_title'].head(100)
```

```python
pattern=r"\s*\-\s|\[|\(|\,|Internship|Training|\/|\sI" # filter Job_title
title=[]
for job in df['Job_title']:
    job=re.sub(r'[^\x00-\x7F]+', '', job) # filter english word because text has some non english word
    job=re.sub('Coders','Programmer',job)
    if re.search(pattern, job ):
        title.append( re.split(pattern,job)[0].strip() )
    else:
        title.append(job.strip())
df['Job_title']=title
df=df[df['Job_title']!='']
```

```python
df['Job_title'].value_counts().reset_index()
```

```python
occu=pd.read_csv('Occupation.csv') 
job_title=df['Job_title'].tolist()
category=occu['Occupation'].tolist()
all_job=job_title+category
all_job
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(1, 3) )
vectors=cv.fit_transform(all_job).toarray()

job_vectors=vectors[:len(job_title)]
category_vectors=vectors[len(job_title):]
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(job_vectors,category_vectors) 
# Match jobs to categories
best_matches = np.argmax(similarity, axis=1)
for i, match in enumerate(best_matches):
    print(job_title[i],"->",category[match])
    job_title[i]=category[match]
df['Job_title']=job_title
```

# create requirement tag 

```python
# create requirement column from description
skills=[]
pattern=r"\b(Requirements?|Qualifications?|Skills?|Must have?|Should have?)\s*[:\-]?\s*(.+?)(?=(?:\n\n|$))"
cnt=0
for des in df['Description']:
    matches = re.findall(pattern, des, re.DOTALL)
    if matches:
        s="".join(match[1] for match in matches)
        s_cleaned = re.sub(r'\n', ' ', s)  # Replace newlines with space
        s_cleaned = re.sub(r'\s+', ' ', s_cleaned).strip()
        skills.append(s_cleaned.lower())
    else:
        skills.append('None')
df['Requirements']=skills
```

#Requirements tag 
-- Data Preprocessing
* Lower case
* Tokenization
* Removing Special character
* Removing stopword and puntuation
* Stemming

```python
import nltk
nltk.download('punkt')
```

```python
import nltk
from nltk.corpus import stopwords
import string

def transform_text(text):
    # Convert text to lowercase
    text = text.lower()
    
    # Replace special characters with spaces
    text = ''.join([char if char.isalnum() or char.isspace() else ' ' for char in text])
    
    # Tokenize the text
    text = nltk.word_tokenize(text)
    
    # Remove stopwords
    text = [word for word in text if word not in stopwords.words('english')]
    
    return " ".join(text)

# Example usage
transform_text("Hi! How %% Are YOU loving problem-solving?")
```

```python
df['Requirements']=df['Requirements'].apply(transform_text)
```

 ### extract skills from requirements

```python
skills_df=pd.read_csv('it_skills_list.csv')
skills_df
```

```python
from sentence_transformers import SentenceTransformer, util

# Load a model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Define skills and job descriptions
skills_list = skills_df['Skills'].tolist()
descriptions_list =df['Requirements'].tolist()

# Generate embeddings for skills
skills_embeddings = model.encode(skills_list)
all_matched_skills = []
# Process each job description
for idx, job_description in enumerate(descriptions_list):
    # Generate embeddings for the current job description
    job_embedding = model.encode(job_description)
    
    # Compute similarities
    similarities = util.cos_sim(job_embedding, skills_embeddings)
    
    # Match skills above the threshold
    threshold = 0.3
    matched_skills = [skills_list[i] for i in range(len(skills_list)) if similarities[0][i] > threshold]
    # Fallback: Add explicitly matched skills
    explicit_matches = [skill for skill in skills_list if re.search(rf'\b{skill}\b', job_description, re.IGNORECASE)]

    for skill in explicit_matches:
        if skill not in matched_skills:
            matched_skills.append(skill)
    # Print results
    print(f"Matched Skills: {matched_skills}\n")
    all_matched_skills.append(matched_skills)
    
df['Skills'] = all_matched_skills
```

```python
df['Skills'] = df['Skills'].apply(lambda x: ', '.join(x))
df
```

# preprocessing of Location

```python
country=[]
division=[]
district=[]
for text in df['Location']:
    entry=re.split(',',text)
    if len(entry)==1:
        cntry=entry[0].strip()
        div='None'
        dist='None'
    elif len(entry)==2:
        cntry=entry[1].strip()
        div=entry[0].strip()
        dist='None'
    else:
        cntry=entry[2].strip()
        div=entry[1].strip()
        dist=entry[0].strip()
    country.append(cntry)
    division.append(div)
    district.append(dist)
df['District']=district
df['Division']=division
df['Country']=country
df.drop(columns='Location',inplace=True)
```
